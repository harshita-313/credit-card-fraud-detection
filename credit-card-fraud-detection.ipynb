```python
# Importing dependencies

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
credit_card_data = pd.read_csv('creditcard.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
credit_card_data.info()
```

```python
credit_card_data.isnull().sum()
```

```python
credit_card_data['Class'].value_counts()

# 0 - Legit transactions
# 1 - Fraud transactions
```

```python
# Separating data for analysis as the ratio of legit data to the fraudulent data is too less.

legit_transactions = credit_card_data[credit_card_data.Class == 0]
fraud_transactions = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit_transactions.shape)
print(fraud_transactions.shape)
```

```python
legit_transactions.Amount.describe()
```

```python
fraud_transactions.Amount.describe()
```

```python
# Compare values for both transactions

credit_card_data.groupby('Class').mean()
```

```python
# Undersampling of legit_transactions as they are around 3 lakhs in no. and only 500 in fraudulent

legit_sample = legit_transactions.sample(n=492)
```

```python
legit_sample.head()
```

```python
legit_sample['Class'].value_counts()
```

```python
# Concatenating 2 dataframes that are legit sample and fraud transactions

new_dataset = pd.concat([legit_sample , fraud_transactions], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset.groupby('Class').mean()
```

```python
# Splitting the data into features and targets

X = new_dataset.drop(['Class'], axis=1)
Y = new_dataset['Class']
```

```python
print (X)
```

```python
print (Y)
```

```python
# Split the data into Training and Testing data

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print (X.shape, X_train.shape, X_test.shape)
```

```python
# Logistic Regression Model Training (Logistic Regression = Binary Classification)

model = LogisticRegression(max_iter=1000)
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print ('Accuracy on Training Data:', training_data_accuracy)
```

```python
model.fit(X_test, Y_test)
```

```python
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print ('Accuracy on Testing Data:', testing_data_accuracy)
```
